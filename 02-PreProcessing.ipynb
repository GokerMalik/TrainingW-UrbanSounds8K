```python
i = 0
fold = path + str(i)
curCatImgs = os.listdir(fold)
    
for img in curCatImgs:
    curImg = matplotlib.image.imread(fold + "/" + img)

    ImgVal = []
    for row in curImg:
        rowVal = []
        for pix in row:
            pixVal = []
            pixNum = pix[0]*0.299 + pix[1]*0.587 + pix[2]*0.114
            pixVal.append(pixNum)
            rowVal.append(pixVal)
        ImgVal.append(rowVal)
        
    L0Img.append(ImgVal)
    L0Res.append(i)
    L0Nam.append(img)

NP0Img = numpy.array(L0Img)
numpy.save("Data/NP0Img", NP0Img)

NP0Res = numpy.array(L0Res)
numpy.save("Data/NP0Res", NP0Res)

NP0Nam = numpy.array(L0Nam)
numpy.save("Data/NP0Nam", NP0Nam)

del NP0Img
del NP0Res
del NP0Nam

del L0Img
del L0Res
del L0Nam

del pixVal
del rowVal
del ImgVal
```
